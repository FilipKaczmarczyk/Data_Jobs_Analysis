```python
import json
import os
import matplotlib.pyplot as plt

current_directory = os.getcwd()
json_file_path = os.path.join(current_directory, '..', 'results', 'top_paying_jobs_results.json')

with open(json_file_path) as f:
    data = json.load(f)

for job in data:
    job['salary_year_avg'] = float(job['salary_year_avg'])

sorted_data = sorted(data, key=lambda x: x['salary_year_avg'])
job_titles = [job['job_title'] for job in sorted_data]
salaries = [job['salary_year_avg'] for job in sorted_data]

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 6))

plt.barh(job_titles, salaries)
plt.xlabel('Average Yearly Salary ($)')
plt.title('Avarage Salary Distribution for Top Paying Data Analyst Jobs in 2023')
plt.xlim(0, max(salaries) * 1.05) 

plt.grid(True, axis='x', linewidth=0.2)

plt.show()
```
