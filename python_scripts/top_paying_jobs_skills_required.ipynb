```python
import json
import os
import matplotlib.pyplot as plt

current_directory = os.getcwd()
json_file_path = os.path.join(current_directory, '..', 'results', 'top_paying_jobs_skills_required_results.json')

with open(json_file_path) as f:
    data = json.load(f)

for skill in data:
    skill['number_of_occurrences'] = float(skill['number_of_occurrences'])

sorted_data = sorted(data, key=lambda x: x['number_of_occurrences'])
skills = [skill['skills'] for skill in sorted_data]
occurrences = [skill['number_of_occurrences'] for skill in sorted_data]

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 6))

plt.barh(skills, occurrences)
plt.xlabel('Number of Occurrences')
plt.title('The Most Common Skills for Top Paying Remote Data Analyst Jobs in 2023')
plt.xlim(0, max(occurrences) * 1.05) 

plt.grid(True, axis='x', linewidth=0.2)

plt.show()
```
